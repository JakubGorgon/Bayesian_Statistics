```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt;import seaborn as sns
```

#### Predicting Email Conversion Rates

```python
x = np.linspace(0, 1, 500)
```

```python
alpha=2
beta=5-alpha
```

```python
pdf_3_2 = stats.beta.pdf(x=x, a=alpha, b=beta)
```

```python
sns.lineplot(x=x, y=pdf_3_2)
plt.xlabel("Conversion Rate") # Out of people who opened email, how many clicked the link ratio (n_click/n_opened)
plt.ylabel("Density")
plt.title("Beta(2,3)")
```

```python
cdf_3_2 = stats.beta.cdf(x=x, a=alpha, b=beta)
sns.lineplot(x=x, y=cdf_3_2)
plt.xlabel("Conversion Rate") # Out of people who opened email, how many clicked the link ratio (n_click/n_opened)
plt.ylabel("Density")
plt.title("CDF Beta(2,3)")
```

```python
sns.lineplot(x=x, y=stats.beta.pdf(x,1,41))
```

```python
alphas = [1, 2, 5, 24]
betas = [41, 80, 200, 976]
x = np.linspace(0, 0.15,500)

for idx, (alpha, beta) in enumerate(zip(alphas, betas)):
    pdf = stats.beta.pdf(x=x, a=alpha, b=beta)
    # plt.figure()
    sns.lineplot(x=x, y=pdf, label=f"Beta({alpha}, {beta})")
    plt.legend()
```

```python
# We choose a 'weak' prior
alpha = 1
beta = 41 

posterior_params = (2+1, 3+41)
posterior_params
```

```python
x = np.linspace(0, 1, 500)
alpha = posterior_params[0]
beta = posterior_params[1]

pdf_posterior_3_44 = stats.beta.pdf(x=x, a=alpha, b=beta)
pdf_likelihood = pdf_3_2

sns.lineplot(x=x, y=pdf_posterior_3_44, label="Posterior (with prior), Beta(3,44)")
sns.lineplot(x=x, y=pdf_likelihood, label="Likehilood (no prior), Beta(2,3)")
```

```python
# Huge impact, even tho it was a weak prior
```

```python
# New data, 100 people open the email, 25 click the link
pdf_likelihood_3_44 = pdf_posterior_3_44

posterior_params = (alpha+25, beta+75)
```

```python

alpha = posterior_params[0]
beta = posterior_params[1]

pdf_posterior_28_119 = stats.beta.pdf(x=x, a=alpha, b=beta)

sns.lineplot(x=x, y=pdf_posterior_28_119, label="Posterior (with prior), Beta(28, 119)")
sns.lineplot(x=x, y=pdf_likelihood_3_44, label="Likelihood, Beta(3,44)")
```

```python
pdf_25_75 = stats.beta.pdf(x=x, a=25, b=75)
sns.lineplot(x=x, y=pdf_25_75, label="Beta(25, 75)")
sns.lineplot(x=x, y=pdf_posterior_28_119, label="Posterior (with prior), Beta(28, 119)")
```
