```python
data = np.array([6,1,3,6,4,5,6,1,2,6])

# is it the loaded die?
```

* H1: Loaded Die
* H2: Fair Die

```python
# we start by working out the bayes factor: P(D|H1)/P(D|H2)
p_6_loaded_die = 1/2
p_not_6_loaded_die = 1/2

p_6_fair_die = 1/6
p_not_6_fair_die = 5/6
```

```python
p_d_given_h1 = np.prod([p_6_loaded_die if x == 6 else p_not_6_loaded_die for x in data])
p_d_given_h2 = np.prod([p_6_fair_die if x == 6 else p_not_6_fair_die for x in data])
```

```python
posterior_odds = (p_h1/p_h2)*bayes_factor
posterior_odds
```
