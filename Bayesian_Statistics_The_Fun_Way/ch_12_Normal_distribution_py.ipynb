```python
import scipy.stats as stats
```

```python
# Define the mean and standard deviation
mean = 20.6
std_dev = 1.62

# Find the probability that a value is less than 18
probability = stats.norm.cdf(18, mean, std_dev)

print(f"The probability is: {probability}")
```

```python
probability = stats.norm.cdf(19,mean, std_dev)
probability_2 = stats.norm.cdf(21,mean, std_dev)

probability_2-probability # PROB OF THE TRUE VALUE BETWEEN 19 and 21
```

```python
probability = stats.norm.cdf((mean-std_dev),mean, std_dev)
probability_2 = stats.norm.cdf((mean+std_dev), mean, std_dev)

probability_2-probability
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and standard deviation
mean = 20.6
std_dev = 1.62

# Generate x values from a range around the mean
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the PDF (probability density function) for each x value
pdf = stats.norm.pdf(x, mean, std_dev)

# # Plot the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Normal Distribution', color='blue')

# # Shade the area where x < 18
x_fill = np.linspace(mean - 4*std_dev, 18, 500)
plt.fill_between(x_fill, stats.norm.pdf(x_fill, mean, std_dev), color='lightblue', label='P(X < 18)')

# # Mark the point x = 18 on the plot
plt.axvline(18, color='red', linestyle='--', label='x = 18')

# # Add labels and legend
plt.title("Normal Distribution with Mean = 20.6, SD = 1.62")
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.legend()

# Show the plot
plt.show()
```

```python
# 2
# temp > 100.4 is fever
values = [100.0, 99.8, 101.0, 100.5, 99.7]

mean = np.mean(values)
std = np.std(values)

probability = stats.norm.cdf(100.4, mean, std)
print(f"{round((1-probability)*100, 2)}% chance of fever")
```
