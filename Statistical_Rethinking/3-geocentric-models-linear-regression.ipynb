#### 1.2 Scientific Model
Height -> Weight <- Unobserved

Weight = f(Height, Unobserved)

"Weight is some function of Height and unobserved stuff"

#### Generative Model
W = Beta*Height + Unobserved

```r
# function to simulate weights of individuals from height
sim_weight = function(H,beta,sd) {
    U = rnorm(length(H), 0, sd)
    W = beta*H + U
    return(W)
}
```

```r
# Generating unformally distributed heights
H = runif(200, 130, 170)
hist(H)
```

```r
W = sim_weight(H=H, beta=0.5, sd=5)
plot(W~H)
```
